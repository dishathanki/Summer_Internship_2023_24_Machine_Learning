# implementing simple diabetes dataset using Sklearn



```python
# just test practice for linar regression !

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
```

```python
# loading dataset to code 

diabetes = datasets.load_diabetes()

dataSet_x = diabetes.data[:,np.newaxis,2]
# print(dataSet_x)

dataSet_x_train = dataSet_x[:-30]
dataSet_x_test = dataSet_x[-30:]
# print(dataSet_x_test)

dataSet_y_train = diabetes.target[:-30]
dataSet_y_test = diabetes.target[-30:]
# print(dataSet_y_test)
```

```python
# creating object for LinearRegression class 

Model = linear_model.LinearRegression();

# fit function is used to train model 

Model.fit(dataSet_x_train,dataSet_y_train)

# here , predict function predicts the value fo y 

dataSet_y_predicted = Model.predict(dataSet_x_test)

# print("Predicted value for y : ")
# print(dataSet_y_predicted)  

#graphical representation using matplotlib 
plt.scatter(dataSet_x_test,dataSet_y_test)
plt.plot(dataSet_x_test,dataSet_y_predicted)
plt.show()
```

# Simple linear Regression Dataset Using Sklearn

```python
#  Importing libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python

# Reading csv File 

data = pd.read_csv('/kaggle/input/linear-regression-dataset/Linear Regression - Sheet1.csv')
data.head()

print(" Dataset shape : ",data.shape)
print(" X shape : ",data['X'].shape)
print(" Y shape : ",data['Y'].shape)
```

```python
#  defining X_train , Y_train , X_test and Y_test Variables for Model 

X = data['X'].values.reshape(-1, 1)
Y = data['Y'].values.reshape(-1, 1)

print(X.shape)
print(Y.shape)

X_train, X_test, Y_train, Y_test = train_test_split(X ,Y , test_size=0.25,random_state=101)

print(" X train shape : ",X_train.shape)
print(" Y train shape : ",Y_train.shape)

print(" X test shape : ",X_test.shape)
print(" Y test shape : ",Y_test.shape)
```

```python
# creating object for LinearRegression 

model = LinearRegression();

model.fit(X_train,Y_train)

Values = model.predict(X_test)
```

```python
# graphical representation of model

plt.scatter(X_test,Y_test)

plt.plot(X_test,Values,color='red')

plt.show()
```
