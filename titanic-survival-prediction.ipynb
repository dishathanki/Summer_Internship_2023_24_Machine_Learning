# Titanic Survival Prediction

```python
# importing libraries for model 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# reading csv file 

data = pd.read_csv('/kaggle/input/titanic-dataset/Titanic-Dataset.csv')

data = data.drop(['Name'],axis=1)
data = data.drop(['Ticket'],axis=1)
data = data.drop(['Cabin'],axis=1)
data = data.drop(['Embarked'],axis=1)


data.replace({'Sex':{'female':1,'male':0}},inplace=True)

data.head()
```

```python
# defining X and Y 

features = data.drop('Survived',axis=1)
target = data['Survived']

scaler =  StandardScaler()
scaler.fit(features)

chngd_data = scaler.transform(features)

features = chngd_data
target = data['Survived']

X_train,X_test,Y_train,Y_test = train_test_split(features,target,test_size=0.2,random_state=2) 
```

```python
# defining Class Logistic_Regression

class LogisiticRegression():
    def __init__(self,learning_rate=0.000005,epochs=1000):
        
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None
        
    def fit(self,X,Y):
        
        self.no_ex,self.no_features = X.shape
        self.weights = np.zeros(self.no_features)
        self.bias = 0
        self.X = X
        self.Y = Y
        
        for i in range(self.epochs):
            self.update_weights()
            
    def update_weights(self):
        
        Y_hat = 1 / (1 + np.exp(-(self.X.dot(self.weights) + self.bias)))
        
        dw = (1/self.no_ex) * np.dot(self.X.T,(Y_hat - self.Y))
        
        db = (1/self.no_ex) * np.sum(Y_hat - self.Y)
        
        self.weights -= self.lr*dw
        self.bias -= self.lr*db
    
    def predict(self,X):
        
        Y_pred = 1 /(1 + np.exp(-(X.dot(self.weights) + self.bias)))
        Y_pred = np.where( Y_pred > 0.5,1,0)
        return Y_pred
```

```python
model = LogisiticRegression()
model.fit(X_train,Y_train)

prediction = model.predict(X_test)

ac = accuracy_score(Y_test,prediction)
print(ac)
```
